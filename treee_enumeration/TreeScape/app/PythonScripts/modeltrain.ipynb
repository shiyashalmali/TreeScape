```python
from ultralytics import YOLO

# Initialize YOLOv8 model with pretrained weights
model = YOLO('yolov8n.pt')

# Train the model
model.train(
    data=r'C:\Users\Dell\Desktop\Enumerate\treedataset\data.yaml',  # Path to data.yaml
    epochs=50,         # Adjust based on time constraints
    batch=4,           # Reduce if CPU/memory is limited
    imgsz=320,         # Image size
    device='cpu'       # Force CPU usage
)
```
